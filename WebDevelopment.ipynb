````python
import flet as ft

def main(page):
  text = ft.Text('Hashzap')
  chat = ft.Column()
  user_name = ft.TextField(label='Enter your name')

  def send_message_tunel(message):
    message_type = message['type']
    if message_type == 'message':
      text_message = message['text']
      user_message = message["user"]
      # Adding message in chat:
      chat.controls.append(ft.Text(f'{user_message}: {text_message}'))
    else:
      user_message = message['user']
      chat.controls.append(ft.Text(f'{user_message} entered the chat', size=12, italic=True, color=ft.colors.ORANGE_500))
      page.update()

      page.pubsub.subscribe(send_message_tunel)

  def send_message(event):
    page.pubsub.send_all({'text': message_field.value, 'user': user_name.value, 'type': 'message'})
    # Cleaning message field:
    message_field.value = ''
    page.update()

    field_message = ft.TextField(label='Enter the message', on_submit=send_message)
    send_button_message = ft.ElevatedButton('Send', on_click=send_message)

  def enter_popup(event):
    page.pubsub.send_all({'user': user_name.value, 'type': 'input'})
    # Adding the chat:
    page.add(chat)
    # Closing popup:
    popup.open = False
    # Remover the 'start chat' button
    page.remove(start_button)
    page.remove(text)
    # Creating the user message field
    # Creating the user send message button
    page.add(ft.Row(
        [message_field, send_button_message]
        ))
    page.update()

    popup = ft.AlertDialog(
        open=False,
        modal=True,
        title=ft.Text('Welcome to Hashzap'),
        content=user_name,
        actions=[ft.ElevatedButton('Enter', on_click=enter_popup)],
    )

    def enter_chat(event):
        page.dialog = popup
        popup.open = True
        page.update()

        start_button = ft.ElevatedButton('start_chat', on_click=enter_chat)

        page.add(text)
        page.add(start_button)

        ft.app(target=main, view=ft.WEB_BROWSER, port=8000)
````
